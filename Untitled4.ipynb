```
import numpy as np 
def gradient_decent(x,y):
  m_curr = 1
  b_curr = 0
  n = len(x)
  iterations = 1000
  learning_rate = 0.001

  for i in range(iterations):
     y_pred = m_curr * x + b_curr
     cost = sum([val **2 for val in (y-y_pred)])
     md  = -2*np.sum(x*(y-y_pred))
     bd  =  -2*np.sum(y-y_pred)
     m_curr = m_curr-learning_rate * md
     b_curr = b_curr-learning_rate * bd
     print("m = {}, c = {} ,iteration = {}, cost {}".format(m_curr,b_curr,i, cost))


x = np.array([0.5,2.3,2.9])
y = np.array([1.4,1.9,3.2])
gradient_decent(x,y)
```
